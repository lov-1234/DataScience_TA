```python
# Q4
df3 = pd.read_csv('./Nightingale.csv')
df3
```

```python
month_order = ["Jan", "Feb", "Mar", "Apr", "May",
               "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

df3['Month'] = pd.Categorical(
    df3['Month'], categories=month_order, ordered=True)
```

```python
# Q4 b)
df3.boxplot(column='Disease', by='Month')
plt.title('Deaths due to Disease by Month')
plt.suptitle('')
plt.xlabel('Month')
plt.ylabel('Number of Deaths')
plt.xticks(rotation=45)
plt.show()
```

```python
df3.boxplot(column='Wounds', by='Month')
plt.title('Deaths due to Wounds by Month')
plt.suptitle('')
plt.xlabel('Month')
plt.ylabel('Number of Deaths')
plt.xticks(rotation=45)
plt.show()
```

```python
df4 = pd.read_csv('Armada.csv')
df4
```

```python
plt.scatter(df4['ships'], df4['artillery'], alpha=0.7)

plt.title('Relationship between Number of Ships and Artillery by Region')
plt.xlabel('Number of Ships')
plt.ylabel('Amount of Artillery')

for i, region in enumerate(df4['Armada']):
    plt.annotate(region, (df4['ships'][i], df4['artillery'][i]),
                 textcoords="offset points", xytext=(5, 5), ha='center')
plt.tight_layout()
plt.grid(True)
plt.show()
```

```python
df4[['artillery', 'ships']].corr()
```

```python
df4['cannon_balls_per_artillery'] = df4['balls'] / \
    df4['artillery']

plt.scatter(df4['artillery'], df4['balls'], alpha=0.7)
plt.title('Relationship between Artillery and Cannon Balls by Region')
plt.xlabel('Amount of Artillery')
plt.ylabel('Amount of Cannon Balls')
plt.grid(True)
plt.show()

df4[['Armada', 'cannon_balls_per_artillery']]
```

```python
df4[['balls', 'artillery']].corr()
```

```python
import seaborn as sns

# Plot pairwise relationships between all numerical variables in the Armada dataset
sns.pairplot(df4.select_dtypes(include=['number']))
plt.suptitle('Pairwise Relationships Between All Numerical Variables', y=1.02)
plt.show()
```
